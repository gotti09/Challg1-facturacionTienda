### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
tienda.columns
```

```python
# Crear la columna Facturacion
tienda['Facturacion'] = tienda['Precio'] + tienda['Costo de envío']

# Calcular facturación total por tienda
facturacion_por_tienda = (
    tienda
    .groupby('Lugar de Compra')['Facturacion']
    .sum()
)

print(facturacion_por_tienda)
```

# 2. Ventas por categoría

```python
categorias_populares = (
    tienda
    .groupby(['Lugar de Compra', 'Categoría del Producto'])
    .size()
    .reset_index(name='Ventas')
)

# Quedarse con la categoría más vendida por tienda
categorias_populares = (
    categorias_populares
    .sort_values(['Lugar de Compra', 'Ventas'], ascending=[True, False])
    .drop_duplicates(subset='Lugar de Compra')
)

print(categorias_populares)
```

```python
productos_populares = (
    tienda
    .groupby(['Lugar de Compra', 'Producto'])
    .size()
    .reset_index(name='Ventas')
)

productos_populares = (
    productos_populares
    .sort_values(['Lugar de Compra', 'Ventas'], ascending=[True, False])
    .drop_duplicates(subset='Lugar de Compra')
)

print(productos_populares)
```

```python
top_categorias = (
    tienda
    .groupby(['Lugar de Compra', 'Categoría del Producto'])
    .size()
    .reset_index(name='Ventas')
    .sort_values(['Lugar de Compra', 'Ventas'], ascending=[True, False])
    .groupby('Lugar de Compra')
    .head(3)
)

print(top_categorias)
```

# 3. Calificación promedio de la tienda


```python
promedio_calificacion = tienda['Calificación'].mean()
print(promedio_calificacion)
```

```python
promedio_calificacion_tienda = (
    tienda
    .groupby('Lugar de Compra')['Calificación']
    .mean()
)

print(promedio_calificacion_tienda)
```

```python
promedio_calificacion_producto = (
    tienda
    .groupby('Producto')['Calificación']
    .mean()
    .sort_values(ascending=False)
)

print(promedio_calificacion_producto)
```

# 4. Productos más y menos vendidos

```python
ventas_producto = (
    tienda
    .groupby(['Lugar de Compra', 'Producto'])
    .size()
    .reset_index(name='Ventas')
)
```

```python
producto_mas_vendido = (
    ventas_producto
    .sort_values(['Lugar de Compra', 'Ventas'], ascending=[True, False])
    .drop_duplicates(subset='Lugar de Compra')
)

print(producto_mas_vendido)
```

```python
producto_menos_vendido = (
    ventas_producto
    .sort_values(['Lugar de Compra', 'Ventas'], ascending=[True, True])
    .drop_duplicates(subset='Lugar de Compra')
)

print(producto_menos_vendido)
```

```python
resumen_productos = (
    ventas_producto
    .sort_values(['Lugar de Compra', 'Ventas'], ascending=[True, False])
    .groupby('Lugar de Compra')
    .agg(
        Producto_mas_vendido=('Producto', 'first'),
        Ventas_max=('Ventas', 'first'),
        Producto_menos_vendido=('Producto', 'last'),
        Ventas_min=('Ventas', 'last')
    )
    .reset_index()
)

print(resumen_productos)
```

# 5. Envío promedio por tienda

```python
costo_promedio_envio = tienda['Costo de envío'].mean()
print(costo_promedio_envio)
```

```python
costo_promedio_envio_tienda = (
    tienda
    .groupby('Lugar de Compra')['Costo de envío']
    .mean()
)

print(costo_promedio_envio_tienda)
```
